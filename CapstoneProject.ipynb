```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns_MA = ['name', 'categories'] + [col for col in dataframe_MA.columns if col.startswith('venue.location.')] + ['venue.id']
dataframe_filtered_MA = dataframe_MA.loc[:, filtered_columns_MA]

nearby_venues_MA = json_normalize(venues_MA) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng', 'venue.location.city']
nearby_venues_MA =nearby_venues_MA.loc[:, filtered_columns]

# filter the category for each row
nearby_venues_MA['venue.categories'] = nearby_venues_MA.apply(get_category_type, axis=1)

# clean columns
nearby_venues_MA.columns = [col.split(".")[-1] for col in nearby_venues_MA.columns]

nearby_venues_MA
```

```python
nearby_venues_MA.groupby(['city','categories']).count()
```

```python
# one hot encoding
MA_onehot = pd.get_dummies(nearby_venues_MA[['categories']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
MA_onehot['city'] = nearby_venues_MA['city'] 

# move neighborhood column to the first column
fixed_columns_MA = [MA_onehot.columns[-1]] + list(MA_onehot.columns[:-1])
MA_onehot = MA_onehot[fixed_columns_MA]

MA_onehot.head()
```

```python
MA_onehot_grouped = MA_onehot.groupby('city').mean().reset_index()
MA_onehot_grouped
```

```python
num_top_venues = 5

for hood in MA_onehot_grouped['city']:
    print("----"+hood+"----")
    temp = MA_onehot_grouped[MA_onehot_grouped['city'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['city']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['city'] = MA_onehot_grouped['city']

for ind in np.arange(MA_onehot_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(MA_onehot_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted
```

# Clustering for Cape Cod

```python
# import k-means from clustering stage
from sklearn.cluster import KMeans
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# set number of clusters
kclusters = 5

lat_lng_MA={'Latitude':dataframe_MA['venue.location.lat'],'Longitude':dataframe_MA['venue.location.lng']}
MA_df_clustering=pd.DataFrame(lat_lng_MA)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(MA_df_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
MA_df_clustering['Name']=dataframe_MA['venue.name']
MA_df_clustering.head()
```

```python
# create map using latitude and longitude values
map_MA = folium.Map(location=[latitude_MA, longitude_MA], zoom_start=10)

# add markers to map
for lat, lng, name, in zip(MA_df_clustering['Latitude'], MA_df_clustering['Longitude'], MA_df_clustering['Name']):
    label = '{}'.format(name)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_MA)  
    
#map_MA
```

```python
MA_df_clustering
```

```python
# add clustering labels
MA_df_clustering.insert(0, 'Cluster Labels', kmeans.labels_)

MA_df_clustering
```

```python
# create map
map_clusters_MA = folium.Map(location=[latitude_MA, longitude_MA], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lng, name, cluster in zip(MA_df_clustering['Latitude'], MA_df_clustering['Longitude'], dataframe_MA['venue.name'], MA_df_clustering['Cluster Labels']):
    label = folium.Popup(str(name) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters_MA)
       
map_clusters_MA
```
